```python
class user:
    def make_turn(self):
        x_move = int(input("\nPlease type the row of your move:"))
        y_move = int(input("Please type the column of your move:"))
        return(x_move, y_move)
    
    def check_user(self):
        check_user_list = list(self.user_mark*self.x)
        diagonal_list = list(np.diagonal(self.board))
        self.inverse = np.transpose(self.board)
        inverse_diagonal = list(np.diagonal(self.inverse))
        if check_user_list == diagonal_list:
            print("You won!")
            return('user')
        for i in range(self.x):
            if check_user_list == list(self.board[i]) or check_user_list == list(self.inverse[i]):
                print("You won!")
                return('user') 
        
class computer:
    def comp_turn(self):
        x_coord = np.random.randint(0,self.x)
        y_coord = np.random.randint(0,self.y)
        return(x_coord, y_coord)
    
    def check_comp(self):
        check_comp_list = list(self.comp_mark*self.x)
        diagonal_list = list(np.diagonal(self.board))
        self.inverse = np.transpose(self.board)
        if check_comp_list == diagonal_list:
            print("Computer won!")
            return('comp')
        for i in range(self.x):
            if check_comp_list == list(self.board[i]) or check_comp_list == list(self.inverse[i]):
                print("Computer won!")
                return('comp')

import numpy as np
import pandas as pd
class tictactoe(user, computer):
    def __init__(self, x=3, y=3):
        self.x = x
        self.y = y

    def choose_player(self):
        self.rand_start = np.random.random()
        if self.rand_start < 0.5:
            print("User makes the first move! You will be 'X'.")
            self.user_mark = 'X'
            self.comp_mark = 'O'
        else:
            print("Computer makes the first move! You will be 'O'.")
            self.user_mark = 'O'
            self.comp_mark = 'X'   
    
    def play(self):
        self.board = np.chararray((self.x, self.y), unicode=True)
        self.choose_player()
        if self.rand_start < 0.5:
            self.play_userstart()
        else:
            self.play_compstart()
    
    def play_userstart(self):
        while '' in self.board:
            userturn = user.make_turn(self)
            while self.board[userturn] != '':
                print("Error, spot already taken. Try again!")
                userturn = user.make_turn(self)
            self.board[userturn] = self.user_mark
            print(self.board)
            if user.check_user(self) == 'user':
                break
            if '' not in self.board:
                break
            compturn = computer.comp_turn(self)
            while self.board[compturn] != '':
                compturn = computer.comp_turn(self)
            self.board[compturn] = self.comp_mark
            print("\nComputer's turn.")
            print(self.board)
            if computer.check_comp(self) == 'comp':
                break
            if '' not in self.board:
                print("It's a tie!!!")
                break
        return("Game over.")
    
    def play_compstart(self):
        while '' in self.board:
            compturn = computer.comp_turn(self)
            while self.board[compturn] != '':
                compturn = computer.comp_turn(self)
            self.board[compturn] = self.comp_mark
            print("\nComputer's turn.")
            print(self.board)
            if computer.check_comp(self) == 'comp':
                break
            if '' not in self.board:
                break
            userturn = user.make_turn(self)
            while self.board[userturn] != '':
                print("Error, spot already taken. Try again!")
                userturn = user.make_turn(self)
            self.board[userturn] = self.user_mark
            print(self.board)
            if user.check_user(self) == 'user':
                break
            if '' not in self.board:
                print("It's a tie!!!")
                break
        print("Game over.")
```
